```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the dataset 'google playstore data' and 'google playstore user reviews'.

```python
df_data = pd.read_csv("../cleansed_googleplaystore.csv")
df_data.head(5)
```

```python
df_review = pd.read_csv("../cleansed_googlereviews.csv")
df_review.head(5)
```

Prepare the dataframe for analysis

```python
#Check for any repeated item in the App column
df_count = df_data['App'].value_counts().to_frame()
df_count = df_count.reset_index()
df_count.rename(columns={'index':'App','App':'value_counts'},inplace=True)
df_count
```

```python
#Take a closer look at the repeated item
df_data[df_data['App']=='ROBLOX']
```

```python
df_data[df_data['App']=='8 Ball Pool']
```

The only difference betweeen the repeated items is the number of reviews.
Let's add a column that shows the number of repeated items .

```python
for i in range(0,len(df_data)):
    df_mean = df_data[df_data['App']==df_data.loc[i,'App']]['Reviews'].mean(axis=0)
    df_data.loc[i,'Reviews'] = df_mean
df_data.head()
```

```python
#check if the 'Reviews' column has changed correctly
df_data[df_data['App']=='8 Ball Pool']
```

```python
#Now we can delete the duplicated apps
df_data.drop_duplicates(subset=['App'],keep='first',inplace=True)
#drop the 'Unnamed: 0' column
df_data.drop(columns=df_data.columns[0], axis=1, inplace=True)
#Reset index number in df_data
df_data.reset_index(drop=True)
```

```python
#Check the number of rows
len(df_data)
```

Get the indicator variables of the specific columns so we can use categorical variables for regression analysis

```python
# get dummy variables of the 'Type' column
dummy_variable_1 = pd.get_dummies(df_data['Type'])
dummy_variable_1.head()
```

```python
# get dummy variables of the 'Content Rating' Column
dummy_variable_2 = pd.get_dummies(df_data['Content Rating'])
dummy_variable_2.head()
```

```python
# rename the column names of dummy_variable_2 for clarity
dummy_variable_2.rename(columns={'Adults only 18+':'content-rating-18+',
                                'Everyone':'content-rating-everyone',
                                'Everyone 10+':'content-rating-10+',
                                'Mature 17+':'content-rating-17+',
                                'Teen':'content-rating-12to18'},inplace=True)
dummy_variable_2.head()
```

```python
# merge the dataframes df_data, dummy_variable_1 and dummy_variable_2
df_data = pd.concat([df_data,dummy_variable_1,dummy_variable_2], axis=1)

#drop the 'Type' and 'Content rating' column from df_data
df_data.drop(['Type','Content Rating'], axis=1, inplace=True)
```

### Data Insights

```python
#Number of apps in each category
df_category = df_data['Category'].value_counts().to_frame()
df_category.rename(columns={'Category':'value_counts'},inplace=True)
df_category
```

Use a bar chart to show the distribution of the number of apps in each category

```python
df_category.plot(kind='bar', figsize=(16,8), color='cornflowerblue')

plt.xlabel("App Categories")
plt.ylabel("Number of Apps")
plt.title("Distribution of the number of apps in each category")
plt.xticks(rotation=45, ha='right')

plt.show()
```

For each category, find out the rating mean.

```python
df_CR = df_data[['Category','Rating']]
df_ratingmean = df_CR.groupby(['Category'],as_index=False).mean()
df_ratingmean.sort_values(by=['Rating'],ascending=False)
```

```python
#Use a boxplot to see the rating range in each app category
plt.figure(figsize=(20,8))
ax = sns.boxplot(x='Category',y='Rating',data=df_data)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12, rotation=40, ha='right')
plt.show()
```

It is observed that the most outliers are appeared in the category 'Family'.

For each category, find out the install sum.

```python
df_CI = df_data[['Category','Installs']]
df_installsum = df_CI.groupby(['Category'],as_index=False).sum()
df_installsum.sort_values(by=['Installs'],ascending=False,inplace=True)
df_installsum.set_index('Category',inplace=True)
df_installsum
```

```python
#Use a bar chart to show the number of installs in each category
df_installsum.plot(kind='bar', figsize=(16,8), color='orangered')

plt.xlabel("App Categories")
plt.ylabel("Number of Installs")
plt.title("Number of installs in each category")
plt.xticks(rotation=45, ha='right')

plt.show()
```

It shows that 'Game' has the most install number, the second is 'Communication'.

**Data Correlations**

```python
#Look at the correlations in df_data
df_data[['Rating','Reviews','Size','Installs','Price']].corr()
```

Only 'Installs' and 'Reviews' have a significant relationships.

```python
#plot a scatter graph to show the relationship between 'Installs' and 'Reviews'
sns.regplot(x='Installs',y='Reviews',data=df_data)
```

```python
#Import linear model from scikit-learn
from sklearn.linear_model import LinearRegression
#create linear regression object
lr = LinearRegression()
#fit the data into the model
X=df_data[['Installs']]
Y=df_data['Reviews']
lr.fit(X,Y)
yhat = lr.predict(X)
```

```python
#show the linear regression equation
print(f'The linear equation is : Reviews = {lr.coef_}*Installs + {lr.intercept_}')
```

```python
#Look at the spread of the residuals
sns.residplot(df_data['Installs'],df_data['Reviews'])
```

The variace of 'Reviews' increase when 'Installs' increases

```python
#Look at the 'Last Updated' column in df_data
df_data['Last Updated'] = pd.to_datetime(df_data['Last Updated'])
df_data.info()
```

```python
#Extract the year in 'Last Updated' column
df_data['year'] = df_data['Last Updated'].dt.year
df_data['year'] = df_data['year'].astype('Int64')
df_data
```

```python
df_data['year'].value_counts()
```

```python
df_data.loc[0,'year']
```

```python
def get_category(year):
    df = df_data[df_data['year']==year]['Category'].value_counts(ascending=False).to_frame()
    return df
```

```python
for i in range(2010,2019):
    print(f'year:{i}')
    print(get_category(i))
    print('\n')
```

It shows that from 2014 to 2018, most of the app made are in the category 'Family'.
